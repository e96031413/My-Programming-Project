# 高雄市停車場資料資訊查詢

```python
import pandas as pd
pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_columns", 1000)
df = pd.read_json('http://parkinglocation.tbkc.gov.tw/getjson.ashx')   #政府公開停車場資料
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  
# 指定默認字形：解決plot不能顯示中文問題
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df.head()
```

# 前鎮區機車車位查詢

```python
near = df[df.行政區 == '前鎮區']
```

## 前鎮區有機車位的地方

```python
new=near[near.機車車位 > 0].sort_index(ascending=False)
```

## 由機車車位高到低排序

```python
new.sort_values(by='機車車位',ascending = False)
```

從上述的資料中可以看出，**夢時代是目前機車車位最多的地方，共有1967位**、草衙道1197位、高雄展覽館857位、台鋁275位

## 苓雅區機車車位查詢

```python
loc2 = df[df.行政區 == '苓雅區']
```

## 苓雅區有機車位的地方

```python
loc2 = loc2[loc2.機車車位>0].sort_index(ascending=False)
```

## 由機車車位高到低排序

```python
loc2.sort_values(by='機車車位',ascending = False)
```

**大統百貨有318個機車車位**，相當的多

## 鳳山區

```python
fs = df[df.行政區 == '鳳山區'].sort_values(by='機車車位',ascending=False)
```

```python
fs = fs[fs.機車車位 > 0]
```

```python
fs_map = fs.drop(labels=['編號','行政區','型式','停車場位置','大型車車位','小型車車位','孕婦及育有六歲以下兒童者停車位','收費標準','營業時間','客服電話'],axis='columns')
fs_map
```
